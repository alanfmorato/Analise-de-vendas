```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pandas.plotting import autocorrelation_plot
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
alucel = pd.read_csv('/content/alucel.csv')
```

```python
alucel.head()
```

```python
alucel.dia = pd.to_datetime(alucel.dia)
```

```python
alucel['aumento'] = alucel['vendas'].diff()
```

```python
alucel['aceleracao'] = alucel['aumento'].diff()
```

```python
plt.figure(figsize=(12,6))
plt.title('Analise vendas 2017-2018')
sns.lineplot(data = alucel, x = 'dia', y = 'vendas');
```

```python
plt.figure(figsize=(12,6))
plt.title('Analise vendas 2017-2018')
sns.lineplot(data = alucel, x = 'dia', y = 'aumento');
```

```python
plt.figure(figsize=(12,6))
plt.title('Analise vendas 2017-2018')
sns.lineplot(data = alucel, x = 'dia', y = 'aceleracao');
```

```python
resultado = seasonal_decompose(alucel['vendas'],  freq=10)
ax = resultado.plot()
```

```python
alucel['media_movel'] = alucel['vendas'].rolling(7).mean()
```

```python
plt.figure(figsize=(12,6))
plt.title('Analise vendas 2017-2018')
sns.lineplot(data = alucel, x = 'dia', y = 'media_movel');
```
